## Предобработка данных

Объединение всех таблиц, используется left join, чтобы не потерять данные (в таблицах internet и phone записей меньше).

```python
data = contract.merge(personal, how='left', on='customer_id')
data = data.merge(internet, how='left', on='customer_id')
data = data.merge(phone, how='left', on='customer_id')
data.sample(10)
```

Исследование пропущенных значений в столбце *total_charges*.

```python
data[data['total_charges'].isnull()]
```

Видно по дате начала пользования услугами, что это новые пользователи. Так как их очень мало  по сравнению с остальной выборкой (11 на 7043), то можно удалить эти записи. 

```python
data = data.dropna(subset='total_charges')
data.info()
```

Целевой признак *actual_client*, в котором 1 будет значить, что клиент продолжает пользоваться услугами, 0 - клиент ушел.

```python
data['actual_client'] = data['end_date'].apply(lambda x: 1 if x == 'No' else 0)
data.sample(5)
```

*No* в столбце *end_date* обозначает, что клиент не прекращал пользоваться услугами. Тогда для того, чтобы привести этот столбец к типу datetime, заменим это значение на дату выгрузки.

```python
data['end_date'] = data['end_date'].replace('No', '2020-02-01')
data['end_date'] = pd.to_datetime(data['end_date'])
data['end_date'].info()
```

Новый столбец *days_using*, в котором указано количество дней пользования услугами.

```python
data['days_using'] = (data['end_date'] - data['begin_date']).dt.days
data.sample(5)
```

```python
data.isna().sum()
```

Так как нет сведений, почему у некоторых клиентов отсутствуют информация об услугах телефонии и интрернета, можно заполнить пропущенные значения в этих столбцах другим значением *Unknown*.

```python
data.loc[:, 'internet_service':'multiple_lines'] = data.loc[:, 'internet_service':'multiple_lines'].fillna('Unknown')
data.isna().sum()
```

Столбцы *customer_id*, *gender*, *begin_date* и *end_date* не несут полезной информации для дальнейшего исследования, поэтому их можно удалить.

```python
data = data.drop(['customer_id', 'gender', 'begin_date', 'end_date'], axis = 1)
data.info()
```

**Вывод:**

* В столбце *total_charges* было принято решение удалить неявные пропущенные значения, их количество очень мало по сравнению со всей выборкой.
* Теперь целевой признак *actual_client*, в котором 1 будет значить, что клиент продолжает пользоваться услугами, 0 - клиент ушел.
* Введен новый столбец days_using, в котором указано количество дней пользования услугами.

## Построение и обучение моделей

### Разбиение на тренировочную и обучающую выборки

```python
data_ohe = data.copy()
data_train, data_test = train_test_split(data_ohe, test_size=0.25, random_state=RANDOM_STATE) 

features_train = data_train.drop(['actual_client'], axis=1)
target_train = data_train['actual_client']
features_test = data_test.drop('actual_client', axis=1)
target_test = data_test['actual_client']

numeric = ['monthly_charges', 'days_using']
categoric = ['type', 'paperless_billing', 'payment_method', 'senior_citizen', 'partner', 
             'dependents', 'online_security', 'online_backup', 'device_protection', 
             'tech_support', 'multiple_lines']

print(f'Размерность признаков - {features_train.shape} и целевого признака - {target_train.shape} обучающей выборки')
print(f'Размерность признаков - {features_test.shape} и целевого признака - {target_test.shape} тестовой выборки')
```

### Использование OHE

```python
encoder_ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)

encoder_ohe.fit(features_train[categoric])
features_train[encoder_ohe.get_feature_names_out()] = encoder_ohe.transform(features_train[categoric])
features_train = features_train.drop(categoric, axis=1)

scaler = StandardScaler()
features_train[numeric] = scaler.fit_transform(features_train[numeric])

features_train.head()
```

```python
features_test[encoder_ohe.get_feature_names_out()] = encoder_ohe.transform(features_test[categoric])
features_test = features_test.drop(categoric, axis=1)

features_test[numeric] = scaler.transform(features_test[numeric])

features_test.head()
```

```python
%%time
xgbmodel = XGBClassifier(random_state=RANDOM_STATE)

param = {'max_depth':[3, 5, 7], 'learning_rate':[.1, .3, .5], 'n_estimators':[400, 500]}

CV = TimeSeriesSplit(n_splits=3).split(features_train)
gscv_xgb = GridSearchCV(xgbmodel, param, cv=CV, scoring='roc_auc')

gscv_xgb.fit(features_train, target_train)
print(f'Лучшие гиперпараметры для XGBClassifier: {gscv_xgb.best_params_}')
```

```python
%%time
catmodel = CatBoostClassifier(random_state=RANDOM_STATE)

param = {'depth':[5, 10], 'learning_rate':np.arange(0.1, 1, 0.2)}

CV = TimeSeriesSplit(n_splits=3).split(features_train)
gscv_cat = GridSearchCV(catmodel, param, cv=CV, scoring='roc_auc')

gscv_cat.fit(features_train, target_train)
print(f'Лучшие гиперпараметры для CatBoostClassifier: {gscv_cat.best_params_}')
```

```python
%%time
trmodel = DecisionTreeClassifier(random_state=RANDOM_STATE) 

param = {'max_depth':[None] + [i for i in range(2, 7)]}

CV = TimeSeriesSplit(n_splits=3).split(features_train)
gscv_tree = GridSearchCV(trmodel, param, cv=CV, scoring='roc_auc')

gscv_tree.fit(features_train, target_train)
print(f'Лучшие гиперпараметры для DecisionTreeClassifier: {gscv_tree.best_params_}')
```

```python
df = [[gscv_tree.best_score_],
      [gscv_cat.best_score_], 
      [gscv_xgb.best_score_]]
model = ["DecisionTree", "CatBoost", "XGBoost"]
pd.DataFrame(data=df, index=model, columns=["Оценка по кросс-валидации"])
```

**Вывод:**
1. Датафрейм был разбит на 2 выборки: тренировочную и тестовую в отношении 3:1 соответственно.
2. С помощью OHE были закодированы признаки, и с GridSearchCV были подобраны лучшие гиперпараметры для 3 моделей.
3. * Лучше всего себя на кросс-валидации показала **CatBoostClassifier** с метрикой *AUC_ROC* ~ 0.88.
   * Совсем немного ей устпупила модель *XGBClassifier* с метрикой *AUC_ROC* ~ 0.87.
   * И хуже всего себя показала *DecisionTreeClassifier* с метрикой *AUC_ROC* ~ 0.78, что даже не входит в порог.
